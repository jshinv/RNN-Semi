```python
T = X_train.shape[1]
i = Input(shape=(T,))

## model 1 - teacher's 0.8826
# model_name = 'model_1.h5'
# x = Embedding(vocab_size, 100)(i)
# x = LSTM(128)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 2 - 0.8949
# model_name = 'model_2.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=16, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = LSTM(32, return_sequences=True)(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 3 - 0.8883
# model_name = 'model_3.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=16, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(32, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 4 - 0.8800
# model_name = 'model_4.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=32, kernel_size=5, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(120, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 5 - 0.8907
# model_name = 'model_5.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=64, kernel_size=3, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(256, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(128, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 6 - 0.8955
# model_name = 'model_6.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=16, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(120, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 7 - 0.8954
# model_name = 'model_7.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=16, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(64, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(32, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 8 - 0.8937
# model_name = 'model_8.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=16, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(128, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(128, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 9 - 0.8883
# model_name = 'model_9.h5'
# x = Embedding(vocab_size, 120)(i)
# x = Conv1D(filters=16, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(128, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Dense(10, activation='sigmoid')(x)
# x = Flatten()(x)
# x = Dense(1, activation='sigmoid')(x)

## model 10 - 0.8926
# model_name = 'model_10.h5'
# x = Embedding(vocab_size, 100)(i)
# x = Dropout(0.2)(x)
# x = Conv1D(filters=32, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(128, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(10, activation='sigmoid')(x)
# x = LSTM(32, return_sequences=False)(x)
# x = Dense(1, activation='sigmoid')(x)

## model 11 - 0.8953
# model_name = 'model_11.h5'
# x = Embedding(vocab_size, 100)(i)
# x = Dropout(0.2)(x)
# x = Conv1D(filters=32, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Dropout(0.2)(x)
# x = LSTM(128, return_sequences=True)(x)
# x = Dropout(0.2)(x)
# x = Dense(10, activation='sigmoid')(x)
# x = Flatten()(x)
# x = Dense(1, activation='sigmoid')(x)

## model 12 - 0.8934
# model_name = 'model_12.h5'
# x = Embedding(vocab_size, 100)(i)
# x = Conv1D(filters=128, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Conv1D(filters=128, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Conv1D(filters=128, kernel_size=4, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=2)(x)
# x = Flatten()(x)
# x = Dense(128, activation='sigmoid')(x)
# x = Dense(1, activation='sigmoid')(x)

## model 13 - 0.8953
# model_name = 'model_13.h5'
# x = Embedding(vocab_size, 100)(i)
# x = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=3)(x)
# x = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=3)(x)
# x = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
# x = MaxPooling1D(pool_size=3)(x)
# x = LSTM(64, return_sequences=True)(x)
# x = Dense(128, activation='sigmoid')(x)
# x = Dense(64, activation='sigmoid')(x)
# x = Flatten()(x)
# x = Dense(1, activation='sigmoid')(x)

## model 14 - 0.8926
model_name = 'model_14.h5'
x = Embedding(vocab_size, 100)(i)
x = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
x = MaxPooling1D(pool_size=3)(x)
x = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
x = MaxPooling1D(pool_size=3)(x)
x = Conv1D(filters=128, kernel_size=3, padding='same', activation='relu')(x)
x = MaxPooling1D(pool_size=3)(x)
x = Dropout(0.2)(x)
x = LSTM(64, return_sequences=True)(x)
x = Dense(128, activation='sigmoid')(x)
x = Dense(64, activation='sigmoid')(x)
x = Flatten()(x)
x = Dense(1, activation='sigmoid')(x)

model = Model(i, x)
model.summary()
```

```python
# 검증 데이터 손실(val_loss)이 증가하면, 과적합 징후
# 검증 데이터 손실이 4회 증가하면 학습을 조기 종료(Early Stopping)
# 검증 데이터의 정확도(val_acc)가 이전보다 좋아질 경우에만 모델을 저장
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4)
mc = ModelCheckpoint(model_name, monitor='val_acc', mode='max', 
                     verbose=1, save_best_only=True)
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
history = model.fit(X_train, y_train, epochs=15, 
                    callbacks=[es, mc], batch_size=60, validation_split=0.2)
```

```python
from tensorflow.keras.models import load_model
model.save(model_name)
```
